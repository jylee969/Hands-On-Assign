```python
import qrcode
```

```python
# This function generates QR code 
def qr_generator(url: str, filename: str = 'img.png'):
    
    # Creates new QRCode object from qrcode we imported 
    QR = qrcode.QRCode()
    
    # Inserts string from the url into empty QR code object in preparation
    # for QR generation
    QR.add_data(url)
   
    # Decides how large QR code needs to be for best QR code size
    # Converts data from information encoded into the QR code (URL address provided)
    # into square pattern 
    QR.make(fit=True)
    
    # Actual image of square pattern or grid to display
    img = QR.make_image()
    
    # Saves the image of QR code
    img.save(filename)
    
    
    

```

```python
if __name__ == "__main__":

    # Asks user for URL so it can be generated into an image
    url = input('Please enter a URL to generate QR code: ').strip()
    
    # Applying function def with the url provided by the user
    qr_generator(url)
```
